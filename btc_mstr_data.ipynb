```python
!pip install ccxt

import ccxt
import pandas as pd
import time

# ✅ Connect to Binance
exchange = ccxt.binance()

# ✅ Fetch order book (top 5 levels)
order_book = exchange.fetch_order_book('BTC/USDT', limit=5)

# ✅ Extract best bid/ask
best_bid = order_book['bids'][0][0]  # price of best bid
best_ask = order_book['asks'][0][0]  # price of best ask

print("Best Bid:", best_bid, "Best Ask:", best_ask)
```

```python
import pandas as pd
from google.colab import files  # for downloading

# ✅ Upload Excel file from your PC into Colab
uploaded = files.upload()  # select btc_mstr_full_data.xlsx

file_path = "btc_mstr_full_data.xlsx"

# ✅ Read BTC and MSTR
btc_df = pd.read_excel(file_path, sheet_name="BTC-USD", skiprows=2)
mstr_df = pd.read_excel(file_path, sheet_name="MSTR", skiprows=2)

btc_df.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]
mstr_df.columns = ["Date", "Close", "High", "Low", "Open", "Volume"]

btc_df['Date'] = pd.to_datetime(btc_df['Date'])
mstr_df['Date'] = pd.to_datetime(mstr_df['Date'])
btc_df.set_index('Date', inplace=True)
mstr_df.set_index('Date', inplace=True)

# ✅ Compute returns
btc_df['Return'] = btc_df['Close'].pct_change()
mstr_df['Return'] = mstr_df['Close'].pct_change()

# ✅ Combine
combined_df = pd.concat(
    [
        btc_df[['Close','Return']].rename(columns={'Close':'BTC_Close','Return':'BTC_Return'}),
        mstr_df[['Close','Return']].rename(columns={'Close':'MSTR_Close','Return':'MSTR_Return'})
    ],
    axis=1
)

# ✅ Save to Excel
output_file = "btc_mstr_returns1.xlsx"
combined_df.to_excel(output_file, sheet_name="Returns")

# ✅ Download to your local drive
#files.download(output_file)
```
